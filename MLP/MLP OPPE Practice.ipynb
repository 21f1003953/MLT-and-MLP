```python
import numpy as np 
import pandas as pd
import itertools
import functools
```

```python
def get_combinations(x,degree):
  return itertools.combinations_with_replacement(x,degree)
```

```python
def compute_new(items):
  return functools.reduce(lambda x,y:x*y,items)
```

```python
def polynomial_features(x,degree):
  if x.ndim==1:
    x=x[:,None]
  xt=np.transpose(x)
  f=[np.ones(len(x))]
  for degree in range(1,degree+1):
    for items in get_combinations(xt,degree):
      f.append(compute_new(items))
  return np.asarray(f).transpose()
```

```python
polynomial_features(np.array([1,2,3]),2)
```
