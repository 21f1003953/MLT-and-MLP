# **DS 1**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import r2_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.preprocessing import Normalizer
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
```

```python
data=pd.read_csv('/content/data_for_large_scale.csv')
random_state=10
```

```python
data.columns
features=['Feature-1', 'Feature-2', 'Feature-3', 'Feature-4', 'Feature-5',
          'Feature-6', 'Feature-7', 'Feature-8', 'Feature-9', 'Feature-10']
data.describe()
```

```python
X=data[features]
y=data.Target
X.head(),y.head()
```

```python
X_array, y_array = X.to_numpy(), y.to_numpy()
X_array, y_array
```

```python
train_X, test_X, train_y, test_y = train_test_split(X_array, y_array, test_size=0.3, random_state=random_state)
```

```python
train_X.shape, test_X.shape
```

```python
train_y.shape, test_y.shape
```

```python
train_X, test_X, train_y, test_y = train_X.reshape((-1,90,10)), test_X.reshape((-1,90,10)), train_y.reshape((-1,90)), test_y.reshape((-1,90))
```

```python
train_X.shape, test_X.shape, train_y.shape, test_y.shape
```

```python
len(train_X)
```

```python
sgd = SGDRegressor(random_state=random_state)
c=[]
for i in range(len(train_X)):
  px = train_X[i]
  py = train_y[i]
  sgd.partial_fit(px,py)
  c.append(sgd.coef_)
```

```python
sgd.intercept_
```

```python
sgd.coef_.round(2)
```

```python
test_X, test_y = test_X.reshape((27000,10)), test_y.reshape((27000,))
```

```python
test_y.shape, test_X.shape
```

```python
y_pred = sgd.predict(test_X)
y_pred
```

```python
c[4][4].round(2)
```

```python
r2_score(test_y, y_pred)
```

# **DS 3**

```python
iris_X, iris_y = load_iris(return_X_y=True)
iris_X, iris_y
```

```python
train_iris_X, test_iris_X, train_iris_y, test_iris_y = train_test_split(iris_X, iris_y,test_size=0.2, random_state = random_state)
train_iris_X.shape, test_iris_X.shape, train_iris_y.shape, test_iris_y.shape
```

```python
tr, ts = Normalizer().fit_transform(train_iris_X), Normalizer().fit_transform(test_iris_X)
tr[:5], ts[:5]
```

```python
for i in [2,3,4]:
  knn=KNeighborsClassifier(n_neighbors=i)
  knn.fit(tr, train_iris_y)
  p = knn.predict(ts)
  print(i,classification_report(test_iris_y, p, output_dict=True)['accuracy'])
  print(i,f1_score(test_iris_y,p, average='weighted').round(2))
```
