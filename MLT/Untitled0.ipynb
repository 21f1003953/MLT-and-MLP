```python
class LinearReg(object):
  def __init__(self,x,y):
    self.x=x
    self.y=y
  
  def predict(self,x,w):
    return np.transpose(self.x)@w
  def  loss(self,x,y,w):
    e=self.predict(x,w)-y
    return 0.5*np.transpose(e)@e
```

```python
def perpred(x,w):
  z=x@w
  return np.sum(np.where(z>0,1,0))
```

```python
perpred(np.array([[1,2],[3,1]]),np.array([[1],[-2]]))
```

```python
def perploss(y,ypred):
  e=np.max(0,-ypred*y)
  return np.sum(e)
```

```python
def graddis(x,y,lr,ep):
  for i in np.arange(0,ep):
    djdw=np.transpose(x)@(x@w-y)
    w=w-lr*djdw
  return w
```
